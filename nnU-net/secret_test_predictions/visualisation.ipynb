```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def find_nifti_files(root_dir):
    """Search recursively for all .nii or .nii.gz files in the given directory."""
    nifti_files = []
    for root, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith(".nii") or file.endswith(".nii.gz"):
                nifti_files.append(os.path.join(root, file))
    return sorted(nifti_files)  # Sorting to maintain order

def load_nifti(file_path):
    """Load a NIfTI (.nii or .nii.gz) file and return the image data as a NumPy array."""
    nii = nib.load(file_path)
    return nii.get_fdata()

def normalize_image(image):
    """Normalize image intensity to [0, 1] for better visualization."""
    return (image - np.min(image)) / (np.max(image) - np.min(image))

def plot_segmentation(image, mask):
    """Displays an interactive visualization of the image and segmentation mask."""
    assert image.shape == mask.shape, f"Shape mismatch: Image {image.shape} vs Mask {mask.shape}"

    fig, axarr = plt.subplots(1, 3, figsize=(15, 5))
    slice_idx = image.shape[2] // 2  # Start from the middle slice

    def update_display(idx):
        axarr[0].cla()
        axarr[1].cla()
        axarr[2].cla()

        # Original Image
        axarr[0].imshow(normalize_image(image[:, :, idx]), cmap="gray")
        axarr[0].set_title(f"Original Image (Slice {idx})")
        axarr[0].axis("off")

        # Segmentation Mask
        axarr[1].imshow(mask[:, :, idx], cmap="jet")  # Show mask separately
        axarr[1].set_title(f"Segmentation Mask (Slice {idx})")
        axarr[1].axis("off")

        # Overlay Image + Mask
        axarr[2].imshow(normalize_image(image[:, :, idx]), cmap="gray")
        axarr[2].imshow(mask[:, :, idx], alpha=0.5, cmap="jet")  # Overlay mask
        axarr[2].set_title(f"Overlay (Slice {idx})")
        axarr[2].axis("off")

        fig.canvas.draw_idle()

    slice_slider = widgets.IntSlider(min=0, max=image.shape[2] - 1, step=1, value=slice_idx)
    widgets.interactive(update_display, idx=slice_slider)

    display(slice_slider)
    update_display(slice_idx)
    plt.show()

# 📝 Set dataset paths
original_images_folder = "../../../dataset/secret_test_set"  # Folder where original images are stored
predicted_masks_folder = "."  # Folder where nnUNet masks are stored

# 🔍 Find all .nii.gz files
original_images = find_nifti_files(original_images_folder)
predicted_masks = find_nifti_files(predicted_masks_folder)

print("Image min/max:", np.min(image_data), np.max(image_data))
print("Mask min/max:", np.min(mask_data), np.max(mask_data))

# 🛑 Ensure we have matching images and masks
if len(original_images) != len(predicted_masks):
    print(f"⚠️ Warning: Found {len(original_images)} images but {len(predicted_masks)} masks!")

# 🎭 Loop through all images and visualize
for img_path, mask_path in zip(original_images, predicted_masks):
    print(f"📂 Visualizing:\n🖼 Image: {img_path}\n🎭 Mask: {mask_path}\n")

    image_data = load_nifti(img_path)
    mask_data = load_nifti(mask_path)

    # 🖼 Show visualization
    plot_segmentation(image_data, mask_data)
```
